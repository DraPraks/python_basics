Yellow pages printing without f-string:

````python
yellow_pages = {'Hakeum':825146987, "Kevin Fe 'popin' Hartono": 5457997746, "Deden Solehudden":245796761}
for name, phone in yellow_pages.items():
  print(name + ' >>> ' + str(phone))
````

the `items()` method is called on the `yellow_pages` dictionary. This method returns a sequence of key-value pairs in the dictionary as tuples.

The for loop then iterates over each tuple in the sequence, and ***assigns the first element of the tuple to the variable `name` and the second element of the tuple to the variable `phone`***. This is known as tuple unpacking.

in the first iteration of the loop, the tuple `('Hakeum', 825146987)` is unpacked into the variables `name` and `phone`, so that ***name is assigned the value 'Hakeum' and phone is assigned the value 825146987***. The print() function then prints the name and phone number of the person.

Yellow pages printing **with** f-string

````python
yellow_pages = {'Hakeum':825146987, "Kevin Fe 'popin' Hartono": 5457997746, "Deden Solehudden":245796761}
for name, phone in yellow_pages.items():
  print(f"{name:10} >>> {phone:10d}")
````

the `10` after `name` and `phone` specifies the minimum number of characters that should be used to display the value of the variable.

If `name` or `phone` is shorter than 10 characters, the remaining characters will be filled with spaces.

The `d` after `phone` is a type specifier that **indicates that the value of phone should be treated as an integer**.

Here are some common list methods in Python:

- `append()`: Adds an element to the end of the list.
- `extend()`: Adds the elements of another list (or any iterable) to the end of the current list.
- `insert()`: Inserts an element at a specified position in the list.
- `remove()`: Removes the first occurrence of an element with a specified value from the list.
- `pop()`: Removes and returns the element at a specified position in the list.
- `index()`: Returns the index of the first element with a specified value.
- `count()`: Returns the number of elements with a specified value.
- `sort()`: Sorts the elements of the list in ascending order.
- `reverse()`: Reverses the order of the elements in the list.


````python
# Create a list of numbers
numbers = [1, 2, 3, 4, 5]
print(numbers)
# Append an element to the end of the list
numbers.append(6)
print(numbers)
# Extend the list with another list
numbers.extend([7, 8, 9])
print(numbers)
# Insert an element at a specific position in the list
numbers.insert(2, 10)
print(numbers)
# Remove an element from the list
numbers.remove(5)
print(numbers)
# Remove and return the last element of the list
last_number = numbers.pop()
print(numbers)
# Get the index of an element in the list
index = numbers.index(4)
print(index)
# Count the number of occurrences of an element in the list
count = numbers.count(3)
print(count)
# Sort the list in ascending order
numbers.sort()
print(numbers)
# Reverse the order of the elements in the list
numbers.reverse()

# Print the final list
print(numbers)
````

# Differences between tuples and lists:
In Python, both lists and tuples are used to store collections of items. Key differences:

1. **Mutability**: Lists are mutable, meaning you can change their content without changing their identity.

You can modify a list by:
- adding
- removing
- changing

tuples are immutable. Once a tuple is created, it cannot be changed.

2. **Usage**: Tuples are typically used for heterogeneous (different) data types and lists for homogeneous (similar) data types¹. Tuples have structure, lists have order. Using this distinction makes code more explicit and understandable¹.

3. **Syntax**: A list is created by placing items (elements) inside square brackets `[]`, separated by commas. A tuple is created by placing items inside parentheses `()`, separated by commas².

Here's an example of how you might use each:

```python
# List of names
names = ['Alice', 'Bob', 'Charlie']

# Tuple representing a book
book = ('The Great Gatsby', 'F. Scott Fitzgerald', 1925)
```

In the list `names`, all elements are similar (they're all names). In the tuple `book`, each element is a different type of data (a title, an author, and a year).

Set operations (Intersection, union, difference, and symmetric difference)

e.g., below.

````python
s1 = {1, 2, 5}
s2 = {2, 3}

print('"s1 & s2": ', s1 & s2) # {2}
print('"s1 | s2": ', s1 | s2) # {1, 2, 3, 5}
print('"s1 - s2": ', s1 - s2) # {1, 5}

print('"(s1 - s2) | (s2 - s1)":', (s1 - s2) | (s2 - s1)) # {1, 3, 5}
print('"s1 ^ s2": ', s1 ^ s2) # {1, 3, 5}
````

````python
s1 = {1, 2, 5}
s2 = {2, 3}
print(s1.intersection(s2)) # {2}, s1 & s2
print(s1.union(s2)) # {1, 2, 3, 5}, s1 | s2
print(s1.difference(s2)) # {1, 5}, s1 -s2
print(s1.symmetric_difference(s2)) # {1, 3, 5}, s1 ^ s2
````

Subsets and Supersets

````python
s1 = {1, 2}
s2 = {1, 2, 3}
s3 = {3, 4}
print(s1 <= s2) # True
print(s2 >= s1) # True
print(s1 <= s3) # False
print(s1 <= s1) # True
````

A **subset** is a set whose all elements are in **another** set. (``issubset()`` or ``<=``)

````python
# Define 2 sets
set_A = {1, 2, 3}
set_B = {1, 2, 3, 4, 5}

# check if set_A is a subset of set_B
print(set_A.issubset(set_B)) # True
print(set_A <= set_B) # True
````

A superset is a set that includes all elements of another set. (```issuperset()``` or ``>=``)

````python
# Define 2 sets
set_A = {1, 2, 3}
set_B = {1, 2, 3, 4, 5}

print(set_B.issuperset(set_A)) # True
print(set_B >= set_A) # True
````

# Dictionaries

- A collection of **key-value** pairs:
  - Each **key** maps to a **value**, which may be any Python object.
- Created using curly braces ``{}`` with comma-separated key-value pairs, or using the ``dict()`` function with a sequence of tuples or keyword arguments.
- Dictionaries can be updated with multiple key-value pairs using the ``update()`` method.
- Cleared of all key-value pairs using the ``clear()`` method.
- Iterated using ``for`` loop or ``items()`` function.

Creating dictionaries:

````python
capitals = {"Indonesia":"Jakarta", "Italy":"Rome", "Germany":"Berlin"}

print(capitals) # {'Indonesia': 'Jakarta', 'Italy': 'Rome', 'Germany': 'Berlin'}
print(type(capitals)) # <class 'dict'>

# Let's make a table cos why not!

# Header
print(f'{"Country":<10} | {"Capitals":<10}')
# <10 specifies that the string should be left-aligned and occupy at least 10 spaces.

# Seperator
print('-' * 23)

# Print each row
for country, capital in capitals.items(): # assigns the key to 'country', value to 'capital'
  print(f'{country:<10} | {capital:<10}')
````

Area of triangle

````python
def area_of_triangle(a, b, c):
    s = (a + b + c) / 2
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    return area

area = area_of_triangle(3, 4, 5)
print("The area of the triangle is:", area)
````

````python
def acronym(phrase): # This line defines a function named acronym that takes one parameter, phrase.
  acronym = "" # This line initializes an empty string, acronym, which will be used to build the acronym.
  words = phrase.split() # This line splits the phrase into a list of words.
  for word in words: # This line starts a for loop that iterates over each word in the words list.
    acronym += word[0].upper() # Takes the first char of the current word (word[0]) and converts it to uppercase.
    # and adds it to the acronym string. The += operator is used to append each letter to the acronym string.
  return acronym # This line returns the final acronym string after all words in the phrase have been processed.

# example usage:
phrase = input('Acronymize: ')
acronym(phrase)
````

# Pandas

## Introduction
- There are 2 main objects in Pandas: ***DataFrame*** and ***Series***.
- A DataFrame is a two-dimensional labeled data structure with columns of potentially different types.
- In Python, DataFrames are provided by the `pandas` library, which is built on top of `numpy`.

A DataFrame is a table containing an array of individual entries, each entry has a certain value. Each entry corresponds to a row and a column.

Let's take this simple example:



````python
pd.DataFrame({'Yes': [50, 20], 'No': [130, 50]}) # DataFrame() func from Numpy
# 'Yes' & 'No' (keys) become the column
````

- pd.DataFrame() or pandas.DataFrame() is a function that creates a DataFrame object.
- The argument added to pd.DataFrame() enclosed in ``{}``, ``'Yes': [50, 20], 'No': [130, 50]`` is a dictionary (key-value pairs).
Sure, let's break down the code `pd.DataFrame({'Yes': [50, 20], 'No': [130, 50]})`:
- `'Yes': [50, 20], 'No': [130, 50]`: These are the key-value pairs in the dictionary:
  - The keys: `'Yes'` and `'No'`
  - values are the lists `[50, 20]` and `[130, 50]`.
  - In the resulting DataFrame, the keys become the column labels, and the lists become the data in the columns.

- `[50, 20]` and `[130, 50]`: These are lists. Lists are ordered collections of items, and they are enclosed in square brackets `[]`.
  - The values in the dictionary are used as the data in df. Each value should be a list (array-like), where each ***order*** of element corresponds to a row in the df.
    - In this case, the value of the lists ``[50, 20]`` and ``[130, 50]``,
    - so the df will have 2 rows:
      - Values of ``50`` from 'Yes' column & ``130`` from 'no' column.
      - ``20`` from 'Yes' and ``50`` from 'No'

So, the entire line of code is creating a DataFrame with two columns labeled 'Yes' and 'No', and the data in those columns is taken from the lists `[50, 20]` and `[130, 50]`.



````python
# df from strings
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']}, index=["Eliza", "Mary"])
# as you can see we added an extra argument 'index' to rename the rows (0, 1, ...)
````

We are using the ``pd.DataFrame()`` constructor to generate these DataFrame objects. The syntax for declaring a new one is a ``dictionary whose keys are the column names`` (Bob and Sue in this example), and whose values are a list of entries. This is the standard way of constructing a new DataFrame, and the one you are most likely to encounter.

## Creating a DataFrame with Numpy Arrays
- You can create a DataFrame from a numpy array. Each row of the array forms a row in the DataFrame, and each column forms a column in the DataFrame.

````python
import numpy as np
import pandas as pd

# Create a 2D numpy array
data = np.array([[1, 2, 3], [4, 5, 6]])

# Create a DataFrame
df = pd.DataFrame(data)

print(df)
````

## Difference between 2D and 3D DataFrames:
### 2D Dataframe:
- Two dimensional labeled data structure with columns of potentially different types.
- Size mutable, size can be changed.
- Basically a table with rows and columns.
#### ***DataFrames can be created from lists, dictionaries, and numpy arrays.***

### 3D DataFrame: MultiIndexed DataFrame/xarray lib:
- A multiIndexed DataFrame is basically a table with more dimensions.
- Xarray is similar to pandas but you can have any number of dimensions.

## Series:
- Sequence of data values; if df is a table, then series is a list.

````python
pd.Series([1, 2, 3, 4, 5])
````

So, a series is basically a single column of a df.

So you can assign row labels to the Series the same way as before, using an index parameter. However, a Series does not have a column name, it only has one overall ``name``:

````python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
````

Overall, think of DataFrames(df) as a bunch of Series "glued together".
